```python
# Import matplotlib's pyplot module
import matplotlib.pyplot as plt

# Step 1: Define data points
x = [1, 2, 3, 4, 5]  # x-axis values (e.g., Months)
y = [10, 20, 25, 30, 50]  # y-axis values (e.g., Sales)

# Step 2: Create a line plot
plt.plot(
    x, y,                     # Input data (x and y must be same length)
    label='Sales Over Time', # Label to appear in the legend
    marker='o',              # Show circular markers at each data point
    color='blue',            # Line color
    linestyle='--'           # Dashed line style
)

# Step 3: Add title to the chart
plt.title('Monthly Sales Data')  # Sets the title above the plot

# Step 4: Label x and y axes
plt.xlabel('Month')              # Label for the horizontal axis
plt.ylabel('Sales in Units')     # Label for the vertical axis

# Step 5: Display the legend
plt.legend()  # Shows what the line represents (based on label='...')

# Step 6: Add grid lines
plt.grid(True)  # Improves readability with background grid

# Step 7: Display the plot
plt.show()  # Renders the plot window
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

# Plot the data
plt.plot(x, y, label='Sales', marker='o')

# Annotate the max value
max_y = max(y)
max_x = x[y.index(max_y)]

plt.annotate(
    'Max Sales: {}'.format(max_y),       # text
    xy=(max_x, max_y),                   # point to highlight
    xytext=(max_x + 0.5, max_y - 10),    # text position
    arrowprops=dict(arrowstyle='->', color='red'),
    fontsize=10,
    color='darkred'
)

# Add other plot elements
plt.title('Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 25, 30, 40])

plt.figure(figsize=(12, 8))  # Optional: increase plot size

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y, label='square', marker='o', color='blue',
         markerfacecolor='red', markeredgecolor='red',
         markersize=10, linewidth=3, linestyle='--')
for i in range(len(x)):
    plt.annotate(f'{x[i]},{y[i]}', (x[i], y[i]), ha='center', va='top', color='green')
plt.title("Subplot 1: x vs y")

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, x, label='normal', marker='x', color='gray',
         markeredgecolor='yellow', markersize=10,
         linewidth=3, linestyle='--')
for i in range(len(x)):
    plt.annotate(f'{x[i]},{x[i]}', (x[i], x[i]), ha='center', va='top', color='green')
plt.title("Subplot 2: x vs x")

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y, label='square', marker='o', color='blue',
         markerfacecolor='red', markeredgecolor='red',
         markersize=10, linewidth=3, linestyle='--')
plt.plot(x, x, label='normal', marker='x', color='gray',
         markeredgecolor='yellow', markersize=10,
         linewidth=3, linestyle='--')
for i in range(len(x)):
    plt.annotate(f'{x[i]},{y[i]}', (x[i], y[i]), ha='center', va='top', color='green')
    plt.annotate(f'{x[i]},{x[i]}', (x[i], x[i]), ha='center', va='bottom', color='blue')
plt.title("Subplot 3: Combined")

# Subplot 4 (Optional)
plt.subplot(2, 2, 4)
plt.plot(x, y - x, label='Diff', marker='s', color='purple',
         markerfacecolor='orange', markeredgecolor='black',
         markersize=10, linewidth=3, linestyle='-.')
for i in range(len(x)):
    plt.annotate(f'{x[i]},{y[i]-x[i]}', (x[i], y[i]-x[i]), ha='center', va='bottom', color='blue')
plt.title("Subplot 4: y - x")

plt.tight_layout()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x ranges
x_wave = np.linspace(0, 2 * np.pi, 100)
x = np.linspace(0.1, 10, 100)
x_tan = np.linspace(-np.pi / 2 + 0.1, np.pi / 2 - 0.1, 100)

# Create 3x3 subplot layout
plt.figure(figsize=(14, 10))

# 1. Sine Wave at (1,1)
plt.subplot(3, 3, 1)
plt.plot(x_wave, np.sin(x_wave), 'bo--', label='sin(x)')
plt.title("Sine Wave")
plt.grid(True)
plt.legend()

# 2. Square Root at (1,2)
plt.subplot(3, 3, 2)
plt.plot(x, np.sqrt(x), 'g^--', label='sqrt(x)')
plt.title("Square Root")
plt.grid(True)
plt.legend()

# 3. Square at (1,3)
plt.subplot(3, 3, 3)
plt.plot(x, x**2, 'rs-', label='x')
plt.title("Square")
plt.grid(True)
plt.legend()

# 4. Cosine Wave at (2,2)
plt.subplot(3, 3, 5)
plt.plot(x_wave, np.cos(x_wave), 'mX--', label='cos(x)')
plt.title("Cosine Wave")
plt.grid(True)
plt.legend()

# 5. Cube at (3,1)
plt.subplot(3, 3, 7)
plt.plot(x, x**3, 'r<-', label='x')
plt.title("Cube")
plt.grid(True)
plt.legend()

# 6. Cube Root at (3,2)
plt.subplot(3, 3, 8)
plt.plot(x, np.cbrt(x), 'c>-', label='cbrt(x)')
plt.title("Cube Root")
plt.grid(True)
plt.legend()

# 7. Tangent Wave at (3,3)
plt.subplot(3, 3, 9)
plt.plot(x_tan, np.tan(x_tan), 'k*--', label='tan(x)')
plt.title("Tangent Wave")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
