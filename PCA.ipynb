```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('iris.csv')

# Check the structure of the dataset
print(df.head())
print(df.columns)

# If the last column is the target class (e.g., 'species'), separate it
X = df.iloc[:, :-1]  # all feature columns
y = df.iloc[:, -1]   # target column (species)

# Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for plotting
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Species'] = y

# Plot the PCA result
plt.figure(figsize=(8, 6))
for species in pca_df['Species'].unique():
    plt.scatter(pca_df[pca_df['Species'] == species]['PC1'],
                pca_df[pca_df['Species'] == species]['PC2'],
                label=species, alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset (from iris.csv)')
plt.legend()
plt.grid(True)
plt.show()
```
